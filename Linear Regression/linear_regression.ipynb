```python
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
np.random.seed(24)
```

## Generating data

```python
# Generate 30 random x values between 1 and 10
x = np.linspace(1, 10, 100)

y = 2 * x + 3 + np.random.normal(0, 1, 100)

# Combine x and y into 2D data points
data = np.column_stack((x, y))
```

## Plotting the data

```python
plt.scatter(data[:, 0], data[:, 1])
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

## Linear Regression model build

```python
class LinearRegression:
    def __init__(self, lr=0.001, w=0.0, b=0.0) -> None:
        self.lr = lr
        self.w = w
        self.b = b
    
    def fit(self, X:np.ndarray, y:np.ndarray, epochs=0):
        sample_size = len(X)

        errors = []
        for _ in range(epochs):
            y_pred = self.w * X + self.b

            # calculating mean squared error
            mse = np.sum(np.square(y - y_pred)) / sample_size

            errors.append(mse)

            # calculate gradients
            w_grad = np.sum(-(2 / sample_size) * X * (y - y_pred))
            b_grad = np.sum(-(2 / sample_size) * (y - y_pred))

            self.w -= self.lr * w_grad
            self.b -= self.lr * b_grad

        return self.w, self.b, errors

    def predict(self, X:np.array) -> np.ndarray:
        return self.w * X + self.b
```

```python
model = LinearRegression(w=0.23, b=0.232, lr=0.001)
epochs = 200
w, b, errors = model.fit(X=data[:, 0], y=data[:, 1], epochs=epochs)
```

## Plotting epoch vs error plot

```python
plt.figure(figsize=(10, 5))
plt.plot(np.arange(1, len(errors) + 1), errors)
plt.xlabel("Epoch")
plt.ylabel("MSE")
plt.show()
```

## Plot the fitted line

```python
plt.figure(figsize=(8, 4))
plt.scatter(data[:, 0], data[:, 1])
plt.plot(data[:, 0], w * data[:, 0] + b, color="orange")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
