```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

```python
# Generate a set of binary data
np.random.seed(0)
m = 100  # Number of samples
true_theta = 0.55  # True probability of getting a 1
data = np.random.binomial(1, true_theta, m)
```

```python
# Define the likelihood function for Bernoulli distribution
def likelihood(theta, data):
    return np.prod(theta**data * (1-theta)**(1-data))
```

```python
# Range of theta values to evaluate
theta_values = np.linspace(0, 1, 100)
likelihood_values = np.array([likelihood(theta, data)
                             for theta in theta_values])
```

```python
# Find the theta that maximizes the likelihood
max_likelihood_index = np.argmax(likelihood_values)
theta_ML = theta_values[max_likelihood_index]
```

```python
# Plot the likelihood function
plt.plot(theta_values, likelihood_values, label='Likelihood')
plt.axvline(theta_ML, color='r', linestyle='--',
            label=f'MLE: theta = {theta_ML:.2f}')
plt.xlabel('Theta')
plt.ylabel('Likelihood')
plt.title('Likelihood Function for Bernoulli Distribution')
plt.legend()
plt.show()

print(f'True theta: {true_theta}')
print(f'Estimated theta (MLE): {theta_ML}')
```
